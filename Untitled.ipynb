## KNN algorithm from scratch

```python
def manhattan_distance(a, b, p=1):
    
  
    dim = len(a)
    a=a.to_numpy()
    b=b.to_numpy()
    
    distance = 0
    
    
    for d in range(dim):
        distance += abs(a[d] - b[d])**p
        
    distance = distance**(1/p)
    
    return distance
```

```python

test_pt = [4.8, 2.7, 2.5, 0.7]

distances = []

for i in data.index:
    distances.append(manhattan_distance(test_pt, data.iloc[i]))
    
df_dists = pd.DataFrame(data=distances, index=data.index, columns=['dist'])
df_dists.head()
```

```python
# Find the 5 nearest neighbors

df_nn = df_dists.sort_values(by=['dist'], axis=0)[:5]
df_nn
```

```python
X_train_numpy_array = X_train.to_numpy()
X_test_numpy_array= X_test.to_numpy()#method 1
y_train_numpy_array=y_train.values#method 2
```

```python
def knn_predict(X_train, X_test, y_train, y_test, k=5):
    y_hat_test = []
    empty_list=[]
    y_predicted=np.array(empty_list)
    for i in range (X_test_numpy_array.shape[0]):
        d=X_train_numpy_array-X_test_numpy_array[i]
        d=d**2
        d=np.sum(d,axis=1)
        d=np.sqrt(d)
        temp_df = pd.DataFrame({'distance': d})
    
    # Sort distances, and only consider the k closest points
    df_nn = temp_df.sort_values(by=['distance'], axis=0)[:k]

    #counter object to track the labels of k closest neighbors
    counter = Counter(y_train[df_nn.index])

    #most common label of all the nearest neighbors
    prediction = counter.most_common()[0][0]

    # Append prediction to output list
    y_hat_test.append(prediction)

    return y_hat_test


```

```python
y_hat_test = knn_predict(X_train, X_test, y_train, y_test, k=5)

print(y_hat_test)
```

```python
def knn(X_train_numpy_array,X_test_numpy_array,y_train_numpy_array,k=3):
  
    empty_list=[]
    y_predicted=np.array(empty_list)
    for i in range (X_test_numpy_array.shape[0]):
        d=X_train_numpy_array-X_test_numpy_array[i]
        d=d**2
        d=np.sum(d,axis=1)
        d=np.sqrt(d)
        temp_df = pd.DataFrame({'distance': d, 'its y train': y_train_numpy_array})
         
        y_hat_test=np.argpartition(d, k)#using introselect algorithm
        b1=np.array([])
        for i in range(k):
            temp_var=temp_df.iloc[y_hat_test[i]]
            b1=np.append(b1,temp_var[1])
        b1=np.mean(b1)
        y_predicted=np.append(y_predicted, b1)
    
   
    return y_predicted
    
```

```python
y_hat_test=knn(X_train_numpy_array,X_test_numpy_array,y_train_numpy_array,5)
print(y_hat_test)
```

```python
print(metrics.r2_score(y_test, y_hat_test))
```
